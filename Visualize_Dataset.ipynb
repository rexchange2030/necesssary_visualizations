# We will be using MNIST dataset for the visualizing 
# Take a look at the dataset

```python
import tensorflow as tf
import math
import matplotlib.pyplot as plt
```

```python
# importing the dataset from keras datasets api
# feel free to explore it at - https://keras.io/api/datasets/fashion_mnist/
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
# it has 10 classes as follows
classes = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankel Boot']
print('the training and test sets consists of following shape --> \n')
print('X_train.shape  --> ', X_train.shape)
print('X_test.shape   --> ', X_test.shape)
print('y_train.shape  --> ',y_train.shape)
print('y_test.shape   --> ', y_test.shape)
```

```python
# Normalizing the data as it is necessary for both visualization and further for feeding the model
X_train, X_test = X_train/255.0, X_test/255.0
```

```python
# plot images
images = 9
from matplotlib import pyplot
for i in range(images):
	# define subplot
	pyplot.subplot(330 + 1 + i)
	# plot raw pixel data
	pyplot.imshow(X_train[i])
# show the figure
pyplot.show()
```
